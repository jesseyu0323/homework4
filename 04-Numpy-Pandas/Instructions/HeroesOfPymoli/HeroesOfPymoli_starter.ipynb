### Heroes Of Pymoli Data Analysis
* Of the 1163 active players, the vast majority are male (84%). There also exists, a smaller, but notable proportion of female players (14%).

* Our peak age demographic falls between 20-24 (44.8%) with secondary groups falling between 15-19 (18.60%) and 25-29 (13.4%).

```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()

```

## Player Count

* Display the total number of players


```python
player_info = purchase_data.loc[:, ["SN", "Age", "Gender"]]
player_info = player_info.drop_duplicates()
number_players = player_info.count()[0]

purchase_data["SN"].value_counts()
player_count = len(purchase_data["SN"].value_counts())
pd.DataFrame([player_count], columns = ["Total Players"])
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = len(purchase_data["Item ID"].unique())
average_purchase_price = purchase_data["Price"].mean()
total_number_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

purchase_analysis = pd.DataFrame({"Number of Unique Items": [unique_item], "Average Price": [average_purchase_price], 
                                  "Number of Purchases": [total_number_purchases], "Total Revenue": [total_revenue]})

purchase_analysis ["Number of Unique Items"] = purchase_analysis["Number of Unique Items"].map("{:,}".format)
purchase_analysis ["Average Price"] = purchase_analysis["Average Price"].map("${:,.2f}".format)
purchase_analysis ["Number of Purchases"] = purchase_analysis["Number of Purchases"].map("{:,}".format)
purchase_analysis ["Total Revenue"] = purchase_analysis["Total Revenue"].map("${:,.2f}".format)

purchase_analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_total = purchase_data["Gender"].value_counts()
gender_percents = gender_total / number_players * 100
gender_demo = pd.DataFrame({"Percentage of Players": gender_percents, "Total Count": gender_total})
gender_demo = round(gender_demo, 2)
gender_demo
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data.groupby(["Gender"]).count()["Purchase ID"]
average_purchase_price = purchase_data.groupby(["Gender"]).mean()["Price"]
total_purchase_price = purchase_data.groupby(["Gender"]).sum()["Price"]
avg_purchase_price_per_pa = total_purchase_price / gender_demo["Total Count"]

purchase_analysis_gender = pd.DataFrame({"Purchase Count": purchase_count,
                                         "Average Purchase Price": average_purchase_price, 
                                         "Total Purchase Value": total_purchase_price, 
                                         "Avg Purchase Total per Person": avg_purchase_price_per_pa})

purchase_analysis_gender ["Purchase Count"] = purchase_analysis_gender["Purchase Count"].map("{:,}".format)
purchase_analysis_gender ["Average Purchase Price"] = purchase_analysis_gender["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_gender ["Total Purchase Value"] = purchase_analysis_gender["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_gender ["Avg Purchase Total per Person"] = purchase_analysis_gender["Avg Purchase Total per Person"].map("${:,.2f}".format)


purchase_analysis_gender
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

player_info["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels = group_names)
age_total = purchase_data.groupby("Age Group")
age_percent = age_total["Purchase ID"].count() / number_players * 100
age_demo = pd.DataFrame({"Percentage of Players": age_percent, "Total Count": age_total["Purchase ID"].count()})
age_demo = round(age_demo, 2)
age_demo.sort_index()
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Group"] = pd.cut(purchase_data["Age"], age_bins, labels=group_names)
age_count = purchase_data.groupby("Age Group").count()["Purchase ID"]
age_average_purchase_price = purchase_data.groupby("Age Group").mean()["Price"]
age_total_purchase_price = purchase_data.groupby("Age Group").sum()["Price"]
age_avg_purchase_price_per_pa = age_total_purchase_price / age_count

purchase_analysis_age = pd.DataFrame({"Purchase Count": age_count,
                                         "Average Purchase Price": age_average_purchase_price, 
                                         "Total Purchase Value": age_total_purchase_price, 
                                         "Avg Purchase Total per Person": age_avg_purchase_price_per_pa})

purchase_analysis_age ["Purchase Count"] = purchase_analysis_age["Purchase Count"].map("{:,}".format)
purchase_analysis_age ["Average Purchase Price"] = purchase_analysis_age["Average Purchase Price"].map("${:,.2f}".format)
purchase_analysis_age ["Total Purchase Value"] = purchase_analysis_age["Total Purchase Value"].map("${:,.2f}".format)
purchase_analysis_age ["Avg Purchase Total per Person"] = purchase_analysis_age["Avg Purchase Total per Person"].map("${:,.2f}".format)

purchase_analysis_age
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
player_purchase_count = purchase_data.groupby("SN").count()["Purchase ID"]
player_average_purchase_price = purchase_data.groupby("SN").mean()["Price"]
player_total_purchase_value = purchase_data.groupby("SN").sum()["Price"]

top_spender = pd.DataFrame({"Purchase Count": player_purchase_count, "Average Purchase Price": player_average_purchase_price, "Total Purchase Value": player_total_purchase_value})
top_spender ["Purchase Count"] = top_spender["Purchase Count"].map("{:,}".format)
top_spender ["Average Purchase Price"] = top_spender["Average Purchase Price"].map("${:,.2f}".format)
top_spender ["Total Purchase Value"] = top_spender["Total Purchase Value"].map("${:,.2f}".format)

top_spender.sort_values("Total Purchase Value", ascending=False).head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
item_purchase_count = purchase_data.groupby(["Item ID", "Item Name"]).count()["Purchase ID"]
item_price = purchase_data.groupby(["Item ID", "Item Name"]).mean()["Price"]
item_total_purchase_value = purchase_data.groupby(["Item ID", "Item Name"]).sum()["Price"]

best_item = pd.DataFrame({"Purchase Count": item_purchase_count, "Item Price": item_price, "Total Purchase Value": item_total_purchase_value})
best_item ["Purchase Count"] = best_item["Purchase Count"].map("{:,}".format)
best_item ["Item Price"] = best_item["Item Price"].map("${:,.2f}".format)
best_item ["Total Purchase Value"] = best_item["Total Purchase Value"].map("${:,.2f}".format)

best_item.sort_values("Purchase Count", ascending = False).head(5)
```
